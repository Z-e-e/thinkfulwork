```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import cv2 
import os
import subprocess
import IPython.display as display
from PIL import Image
import glob
```

```python
import tensorflow as tf
import keras
from keras.models import Sequential 
from keras.layers import Conv2D, MaxPooling2D #extract features, reduce size
from keras import backend as K
from keras.layers import Activation, Dropout, Flatten, Dense #when to activate, NN doesn't overfit, conver to 1d, output layer
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
```

```python
img_width, img_height = 250, 250

if K.image_data_format() == 'channels_first':
    input_shape=(3,img_width, img_height)
else:
    input_shape=(img_width,img_height, 3)
```

```python
#flow_from_directory method
#defines tarain data set
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        validation_split=0.2)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
#loads images in training set
print ('Training Set:')
train_generator=train_datagen.flow_from_directory(
directory="C:/Users/zalth/Desktop/final_cap/"+"/Images_w_class/",
target_size=(img_width,img_height),
class_mode = "categorical",
batch_size=32,
subset='training')
print ('_______________________________________________________')
print ('Validation Set:')
#loads images to validation set
validation_generator=train_datagen.flow_from_directory(
directory="C:/Users/zalth/Desktop/final_cap/"+"/Images_w_class/",
target_size=(img_width,img_height),
class_mode = "categorical",
batch_size=32,
subset='validation')
```

```python
print(len(validation_generator))
print(len(validation_generator[0]))
print(validation_generator[0][0].shape)
```

```python
#CNN

model = Sequential()
model.add(Conv2D(32,(3,3), input_shape=input_shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.summary()

model.add(Conv2D(32,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2))) #reduce the size to half 

model.add(Conv2D(64,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model(Activation('softmax'))

model.summary()

model.compile(loss='categorical_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])

model.fit_generator(
        train_generator,
        steps_per_epoch=2000,
        epochs=50,
        validation_data=validation_generator,
        validation_steps=800)
```
